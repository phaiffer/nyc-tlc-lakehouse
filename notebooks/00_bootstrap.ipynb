```python
from pathlib import Path

YEAR = 2024
MONTH = 1

RAW_DIR = Path("../data/raw")
RAW_DIR.mkdir(parents=True, exist_ok=True)

MM = f"{MONTH:02d}"
RAW_FILE = f"yellow_tripdata_{YEAR}-{MM}.parquet"
RAW_PATH = RAW_DIR / RAW_FILE

print("Run order: CONFIG -> Download -> Preview")
print("RAW_PATH:", RAW_PATH)
```

```python
import urllib.request

DOWNLOAD_URL = f"https://d37ci6vzurychx.cloudfront.net/trip-data/{RAW_FILE}"

if not RAW_PATH.exists():
    print("Downloading:", DOWNLOAD_URL)
    urllib.request.urlretrieve(DOWNLOAD_URL, RAW_PATH)
else:
    print("Already downloaded:", RAW_PATH)

if not RAW_PATH.exists():
    raise FileNotFoundError(f"Raw parquet was not found after download attempt: {RAW_PATH}")

print("Raw parquet ready:", RAW_PATH)
```

```python
import pyarrow.parquet as pq

table = pq.read_table(RAW_PATH)

print("Schema:")
print(table.schema)

head_preview = table.slice(0, 5)
print("Head rows (pyarrow):")
print(head_preview)
```

```python
print("total_rows:", table.num_rows)
print("total_columns:", table.num_columns)
```
